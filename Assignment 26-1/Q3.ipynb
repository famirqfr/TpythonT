```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv.imread('Image/a.jpg')
result = np.zeros(image.shape, image.dtype)
```

```python
for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        for c in range(image.shape[2]):
            result[y,x,c] = np.clip(1.5*image[y,x,c] + 50, 0, 255)
```

```python
plt.imshow(image)
```

```python
plt.imshow(result)
```
