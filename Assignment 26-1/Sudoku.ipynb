```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Image/sudoku.tif' , cv2.IMREAD_GRAYSCALE)
plt.imshow(img)
```

```python

optimized_img = cv2.equalizeHist(img)
(thresh , optimized_img)= cv2.threshold(optimized_img , 128 ,255 , cv2.THRESH_BINARY | cv2.THRESH_OTSU)

x = [] 
y = [] 
for i in range(optimized_img.shape[0]):
    for j in range(optimized_img.shape[1]):
        if optimized_img[i][j] == 0:
            x.append(j)
            y.append(i)

plt.imshow(optimized_img)
```

```python
minx=min(x) 
maxx=max(x)
miny=min(y) 
maxy=max(y)
```

```python
result = optimized_img[miny:maxy , minx:maxx]
plt.imshow(result)
```

```python
rows, cols = result.shape
```

```python
l = rows//9
for i in range(0, rows, l):
    for j in range(0, cols, l):
        cell = result[int(i)+15:int(i+l)-15, int(j)+15:int(j+l)-15]
        if np.average(cell)<15:
            cv2.imwrite(f'output{i}{j}.jpg', cell)
```
